```python
chat = []
for saved_chat in ('meeting_saved_chat_2_1.txt', 'meeting_saved_chat_2_2.txt'):
    with open(os.path.join(CHAT_FOLDER, saved_chat), 'rt', encoding='UTF-8') as iput:
        chat += iput.readlines()
chat = [line.strip() for line in chat]
```

```python
chat_pattern = '(?P<time>\d{2}:\d{2}:\d{2}) От  (?P<user>.+)  кому  .*:'
```

```python
time_list = []
user_list = []
message_list = []
message = ''

for line in chat:
    match = re.match(chat_pattern, line)
    if match is None:
        if message and line:
            message += '\n'
        message += line
        continue
        
    if time_list:
        message_list.append(message)
        message = ''
    
    time_list.append(match.group('time'))
    user_list.append(match.group('user'))

message_list.append(message)
```
